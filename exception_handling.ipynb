1 . What is the difference between interpreted and compiled languages?

**Answer:**
In interpreted languages, the code is executed line by line by an interpreter (e.g., Python). In compiled languages, the entire program is translated into machine code before execution (e.g., C, C++).

* **Interpreted:** Slower execution, but easier to debug.
* **Compiled:** Faster execution, but requires compilation before running.

2 . What is exception handling in Python?

**Answer:**
Exception handling is a way to handle runtime errors in Python using `try`, `except`, `finally`, and `else` blocks. It prevents the program from crashing and allows us to deal with errors properly.

3. What is the purpose of the finally block in exception handling?

**Answer:**
The `finally` block is used to write code that must run whether an exception occurs or not. For example, closing a file after reading.

4. What is logging in Python?

**Answer:**
Logging in Python is a way to record events or messages (like errors, warnings, or program status) while the program runs. It helps in debugging and monitoring.

5. What is the significance of the **del** method in Python?

**Answer:**
The `__del__` method is a destructor in Python. It is called automatically when an object is deleted or goes out of scope. It is often used to release resources.

6. What is the difference between import and from ... import in Python?

**Answer:**

* `import module` → You must use the module name to access functions (`math.sqrt(16)`).
* `from module import function` → You can use the function directly (`sqrt(16)`).

25. Why is it important to use multithreading in certain applications? (with example)

**Answer:**
Multithreading is important when tasks can run concurrently, like downloading files or handling multiple users in a chat app. It improves efficiency in I/O-bound tasks.

**Example:**

```
import threading
import time

def task(name):
    print(f"Task {name} started")
    time.sleep(2)
    print(f"Task {name} finished")

t1 = threading.Thread(target=task, args=("A",))
t2 = threading.Thread(target=task, args=("B",))

t1.start()
t2.start()
```

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("rotate.log", maxBytes=1024*1024, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO)

logging.info("This is a log entry")
```
